```python
import pandas as pd
import numpy as np
```

```python
#reading the data
df = pd.read_csv(r'C:\\Users\dell\Downloads\NIFTY.csv', thousands=',' )
df.head()
```

```python
df=df.replace('-', np.nan)
df.dtypes
```

```python
df1 = df[['OI_call', 'STRIKE PRICE','OI']]
df1.head()
```

```python
for i in df1.columns:
    df1[i] = pd.to_numeric(df1[i].replace(',', '',regex=True))
```

```python
Call = df1.nlargest(3, ['OI_call'])
print(Call[['OI_call','STRIKE PRICE']])

Put = df1.nlargest(3, ['OI'])
print(Put[['OI','STRIKE PRICE']])
```
