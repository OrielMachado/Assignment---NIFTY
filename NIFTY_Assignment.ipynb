```python
import pandas as pd
import numpy as np
```

```python
#reading the data

df = pd.read_csv(r'C:\\Users\dell\Downloads\option-chain-ED-NIFTY-26-Aug-2021.csv', thousands=',' )
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df1 = df[['level_1', 'level_11', 'Unnamed: 1']]
df1.head()
```

```python
df1=df1.replace('-', np.nan)

df1.columns = ['OI_call', 'StrikePrice','OI_put']
df1.drop(index=0,axis=0, inplace=True)

df1.head()
```

```python
df1.dtypes
```

```python
for i in df1.columns:
    df1[i] = pd.to_numeric(df1[i].replace(',', '',regex=True))
df1.dtypes    
```

```python
Call = df1.nlargest(3, ['OI_call'])
print(Call[['OI_call','StrikePrice']])
```

```python
Put = df1.nlargest(3, ['OI_put'])
print(Put[['OI_put','StrikePrice']])
```
